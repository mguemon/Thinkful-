```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
%matplotlib inline
plt.rcParams['figure.figsize'] = (20.0, 10.0)
```

```python
#import data frame and view head
df = pd.read_csv('SoccerLEagues.csv')
```

The Data come from Kaggle Data: [Soccer Data](https://www.kaggle.com/drgilermo/home-advantage-in-soccer-and-basketball/data)  
This Data has the different Soccer leagues from different counries and gives their Home Wins, Home Loss, Away Wins and Away Loss for a particular year.
The years are from 2009 to 2016.

I would like to see if there is a correlation between the Home Wins, Away Wins for the different teams/country.
And if there is one what may that be.

```python
#view head of Data
df.head()
```

```python
#This data gives us the countryies with the most Home Wins.
df[df['HomeWins'] == df['HomeWins'].max()].head()
```

```python
#This data gives us the countryies with the most Away Wins.
df[df['AwayWins'] == df['AwayWins'].max()].head()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.hist(df['HomeWins'], bins=19, color='red', alpha = 0.5)
plt.hist(df['HomeLoss'], bins=17, color='blue', alpha = 0.5)
plt.ylabel('Total Home Games')
plt.xlabel('Home Games Won')


plt.subplot(1, 2, 2)
plt.hist(df['HomeWins'], bins=19, color='red', alpha = 0.5)
plt.hist(df['AwayWins'], bins=16, color='green', alpha = 0.5)
plt.ylabel('Total Away Games')
plt.xlabel('Home Games Won')

plt.show()
```

The 1st Figure is a histogram that shows a difference between Home Wins (red) and Home Loss (blue).  
The purple represent where they overlap.

The 2nd Figure is a histogram that shows a difference between Home Wins (red) and Away Wins (green)
The Dark Green represent where they overlap. 

This shows that the two populations are different.


```python
df.boxplot(column = 'HomeWins', by = 'Country' )
plt.xticks(rotation=90);
```

From this boxplot, we can clearly see the mean Home Wins per country.  
The mean is different from one country to the next.
We can see that countries like Indonesia and Nigeria have on average the most Home Wins. 
While countries like Congo and Papau have on average the least Home Wins on Average.

```python
df.boxplot(column = 'AwayWins', by = 'Country' )
plt.xticks(rotation=90);
```

From this boxplot, we can clearly see the mean Away Wins per country  
It is clear that most countries Away Wins are on average around 4 games.  
But for countries like Estonia, Lithuana that are bit higher with a larger upper and lower quartiles.
While countries like Congo, Guatamala, Nigeria and Papau have the least Away Wins on average.
