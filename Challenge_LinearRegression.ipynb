```python
# Making dummies.
# Also creating a data frame to hold our features.
features = df
display(features.head())
```

```python
### Create the new feature, (Population)^2
features['Population_sqrd'] = df['Population'] **2
#where values greater than 0 are coded 1, and values equal to 0 are coded 0.

features['Robbery_cat'] = np.where(df['Robbery']>0, 1, 0)
features['Murder_cat'] = np.where(df['Murder and\nnonnegligent\nmanslaughter']>0, 1, 0)
```

```python


df_f = features[['City', 'Property\ncrime','Population', 'Population_sqrd','Murder and\nnonnegligent\nmanslaughter',
                 'Murder_cat','Robbery','Robbery_cat']]

display(df_f.head())


# Instantiate and fit our model.
regr = linear_model.LinearRegression()
Y = df_f['Property\ncrime']
X = df_f[['Population', 'Population_sqrd','Murder_cat','Robbery_cat']]
regr.fit(X, Y)

# Inspect the results.
print('\nCoefficients: \n', regr.coef_)
print('\nIntercept: \n', regr.intercept_)
print('\nR-squared:')
print(regr.score(X, Y))
```

2- multivariate normality

```python
# Extract predicted values.
predicted = regr.predict(X).ravel()
actual = df_f['Property\ncrime']

# Calculate the error, also called the residual.
residual = actual - predicted


plt.hist(residual)
plt.title('Residual counts')
plt.xlabel('Residual')
plt.ylabel('Count')
plt.show()
```

```python
# Transformation causes error. Check for zeroes in property_crime (dependent variable).
display(df_f[df_f['Property\ncrime'] == 0])
```

```python
# Drop rows with zeroes.
df_f1 = df_f[df_f['Property\ncrime'] != 0]
```

```python
# Rerun the model with log transformation on the dependent variable since removed outlier

Y1 = np.log(df_f1['Property\ncrime'])
X1 = df_f1[['Population', 'Population_sqrd','Murder_cat','Robbery_cat']]
regr.fit(X1, Y1)

# Inspect the results.
print('\nCoefficients: \n', regr.coef_)
print('\nIntercept: \n', regr.intercept_)
print('\nR-squared:')
print(regr.score(X1, Y1))
```

```python
# Verify residual distribution.

# Extract predicted values.
predicted1 = regr.predict(X1)
actual1 = Y1
# Calculate the error, also called the residual.
residual1 = actual1 - predicted1

plt.hist(residual1)
plt.title('Residual counts')
plt.xlabel('Residual')
plt.ylabel('Count')
plt.show()
```

```python
plt.scatter(predicted, residual, alpha=0.6 )
plt.xlabel('Predicted')
plt.ylabel('Residual')
plt.axhline(y=0)
plt.title('Residual vs. Predicted')
plt.show()
```
