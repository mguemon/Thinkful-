```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('weatherdata.csv')
df.head() #Not using Print statement so that it displays all in one line for all tables 
```

## Choose one variable and plot that variable four different ways.

```python
plt.hist(df['precipitation'], color='red',  alpha=.5,label='Precipitation', bins=20)
plt.show()
```

```python
xp = df['precipitation']
ax = sns.distplot(xp, bins=20 )
plt.show()
```

```python
df.plot.line(x=df.index, y='precipitation')
```

```python
df['precipitation'].hist(bins=20)
```

## Choose two continuous variables, and plot them three different ways..


```python
#sns.lmplot(x="temperature", y="humidity", data=df, x_jitter=.05);
g = sns.jointplot(x="temperature", y="humidity", data=df, kind="reg");
plt.show()
```

```python
sns.jointplot(x='temperature',y='humidity',data=df,kind='hex')
```

## Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
ax = sns.pointplot(x="year", y="precipitation", data=df)
plt.show()
```

```python
ax = sns.boxplot(x="year", y="precipitation", data=df)
plt.show()
```

```python
sns.violinplot(x="year", y="precipitation",data=df,split=True)
```

```python
sns.stripplot(x="year", y="precipitation",data=df, jitter=True)
```

```python
sns.lvplot(x="year", y="precipitation",data=df)
```

```python
sns.countplot(x="year", data=df)
```
