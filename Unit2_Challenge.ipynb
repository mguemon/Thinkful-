```python
df.boxplot(column = 'HomeWins', by = 'Year' )
plt.xticks(rotation=90);
```

From this Boxplot, we can see that in 2016,   
the teams did not have a lot of Home Wins.

```python
df.boxplot(column = 'AwayWins', by = 'Year' )
plt.xticks(rotation=90);
```

From this Boxplot, we can see that in 2012, 2013 and 2014 the teams were consistent in the amount of AWay Wins they got.

```python
plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 1)
df.groupby('HomeLoss').HomeLoss.mean().plot(kind = "bar")

plt.subplot(1, 2, 2)
df.groupby('AwayWins').AwayWins.mean().plot(kind = "bar")

plt.tight_layout()
plt.show()
```

From these bar charts, we see that the Home Loss and Away Wins mean increase over the amount of games played.
